# AB Schleifensteuerung – Übungen zur for-Schleife 


```python
capital = float(input("Das Startkapital: "))
interest = float(input("Der Zinssatz: "))


start_capital = capital
double_capital = capital * 2
years = 0

while start_capital < double_capital:
    start_capital += start_capital * interest
    years += 1

print(f"Es dauert {years} Jahre, bis das Kapital {double_capital:.2f} Euro beträgt.")
```

# Wiederholungsstrukturen - Übungen zur zählergesteuerten Schleife

### Aufgabe 1

```python

anzahl_schueler = int(input("Anzahl der Schüler/innen: "))
summe_alter = 0

for i in range(1, anzahl_schueler + 1):
    alter = int(input(f"Alter von Schüler/in {i}: "))
    summe_alter += alter

durchschnittsalter = summe_alter / anzahl_schueler
print(f"Das Durchschnittsalter der Klasse beträgt: {durchschnittsalter:.2f} Jahre")
```
